```python

"""
1. Display Menu
"""
def menu():
    print("\nInventory Management System")
    print("1. Add a new product")
    print("2. Update product details")
    print("3. Add a new supplier")
    print("4. Place an order")
    print("5. View inventory")
    print("6. Generate reports")
    print("7. Exit")

"""
2. Read Data
"""
def read_data(x): #read the files to separate them by comma
    f = open(x, "r")
    return [line.strip().split(',') for line in f.readlines()]

"""
3. Add Products
"""
def add_products():
    f = open("products.txt", "a") #open the product file
    product_id=input("Enter Product ID:") #Insert product ID
    while not product_id: # Check if the input is empty
        print("Product ID cannot be empty. Please enter a valid Product ID.")
        product_id = input("Enter Product ID:").strip() # Prompt again
    
    product_name=input("Enter Product Name:") #Insert product name
    while not product_name: # Check if the input is empty
        print("Product Name cannot be empty. Please enter a valid Product Name.")
        product_name = input("Enter Product Name:").strip() # Prompt again
    
    product_description = input("Enter Product Description(If multiple description use / to seperate):") #Insert product description
    while not product_description: # Check if the input is empty
        print(" Product Description cannot be empty. Please enter a valid Product Description.")
        product_description = input("Enter Product Description:").strip() # Prompt again
    
    product_price=float(input("Enter Product Price(In RM):")) #Insert the data price
    while not product_price>0: # Check if the price is a negative number
        print("Error Negative Number: Enter A Positive Positive Number")
        product_price=float(input("Enter Product Price(In RM):"))
    
    product_stock=input("Enter The Stock Quantity Of The Product(in numbers)")
    while not product_stock: # Check if the input is empty
        print("The Stock Quantity Of The Product cannot be empty. Please enter a valid Stock Quantity Of The Product")
        product_stock = input("Enter The Stock Quantity Of The Product:").strip() # Prompt again
    
    products=read_data("products.txt")#read the data from product file
    for product in products: #check if the product id is in the file
        if product[0] == product_id:
            print("Product ID already exists. Please use a unique ID.")
            return
        
    f.write(product_id+","+product_name+","+product_description+",RM"+str(product_price)+","+str(product_stock)+"\n") #write the new product into the file
    print("Product Added Succesfully")
    f.close
    
"""
4. Update Products
"""
def update_products():
    product_id = input("Enter Product ID To Update: ")#User insert the product id that they want to edit
    products = read_data("products.txt")#read the data from product file
    updated = False # Flag to check if the product was found and updated
    
    for product in products:
        if product[0] == product_id: # Match the product id in the file
            product[1] = input("Enter New Product Name: ") #edit the new Product Name
            product[2] = input("Enter New Product Description (If multiple descriptions use / to separate): ") #edit the new description
        product_price = float(input("Enter New Product Price: ")) #New price
        
        if product_price<0:# Check if the price is a negative number
            print("Error Negative Number: Enter A Positive Positive Number")
            return
        product[3] = product_price
        product[4]=input("Enter The Stock Quantity Of The Product")
        updated = True # Set the flag to True
        print("Product Updated Successfully")
        break # Exit the loop aher updating
    
    if not updated: #if the product id was not found in the file
        print("Product not found.")
        return
        
    f=open("products.txt","w")
    for product in products: #edit the product details in the file
        f.write(product[0] + "," + product[1] + "," + product[2] + ",RM" + str(product[3])+","+str(product_stock)+"\n")
        
    f.close
    
"""
5. Show Supplier
"""
import os
### Ensure the suppliers.txt file exists and Show the supplier informations existed in the file
def show_supplier():
    """
    Ensure the file exists, and will Create the file if it does not exist
    """
    if not os.path.exists('suppliers.txt'):
        with open('suppliers.txt', 'w') as sup_file: 
            pass

    """
    Show the supplier informations existed in suppliers.txt file
    """
    # If Invalid input, still continuing asking for Valid input. 
    # Only stops when supplier informations is displayed or requested by the user.
    while True:  
        sup_show = input("\nDo you want to display supplier informations existed in suppliers.txt file(yes/no): ")
        sup_show = sup_show.lower() # lowercase
        sup_show = sup_show.replace(" ", "") # Remove whitespaces
        if sup_show == "yes": # Read the supplier informations from the file and display them
            with open('suppliers.txt', 'r') as sup_file:
                sup_lines = sup_file.readlines() 
                # reads the entire content of the file and returns it as a list, 
                # each element in the list corresponds to one line from the file.
                
                # Check if the file is empty
                if not sup_lines:
                    print("No suppliers found in the file.")
                    break  # Exit the function if no suppliers are found

                ### Extract the supplier ID from a line of text
                def extract_supplier_id(line):
                    return int(line.split(',')[0].split(":")[1].strip()) # Extract and return the supplier ID (as an integer)  
                    '''
                    e.g.                                            ID: 01, Name: Supplier1, Contact: 1234567890
                    #line.split(',')                              ['ID: 01', ' Name: Supplier1', ' Contact: 1234567890']
                    #line.split(',')[0]                            'ID: 01'
                    #line.split(',')[0].split(":")             ['ID', ' 01']
                    #line.split(',')[0].split(":")[1]                 ' 01'
                    #line.split(',')[0].split(":")[1].strip()          '01'
                    #int(line.split(',')[0].split(":")[1].strip())       1
                    '''
                
                # Sort the suppliers based on the ID
                sup_lines.sort(key=extract_supplier_id) # key: extract the value of ID(already in int) to compare for each element in sup_lines
                
                # Sort the suppliers based on the ID in the suppliers.txt file
                with open('suppliers.txt', 'w') as sup_file:
                    sup_file.writelines(sup_lines)  # Write the sorted lines to the suppliers.txt file  

                # Print each line of the file, which contains Supplier details
                print("\nList of Suppliers:") 
                for line in sup_lines: 
                    print(line.strip())  # .strip() removes any leading/trailing whitespace (including newline)
            break # Break the loop
                    

        elif sup_show == "no": # Break the loop
            break

        else: # Invalid Option
            print("Error: Invalid Option. Please try again(yes/no).")
            continue # Continue asking for a Valid Option
        
"""
6. Add Suppliers
"""
### Add a new supplier: Allow the user to add a new supplier to the system.
def add_supplier(): 
    """
    Ensure the suppliers.txt file exists and Show the supplier informations existed in the file
    """  
    show_supplier()

    """
    Read the existing Supplier IDs from the file
    """        
    with open('suppliers.txt', 'r') as sup_file:
        existing_id = [line.split(',')[0].split(":")[1].strip() for line in sup_file]
        #This process is repeated for each line in the file, and the result is a list of all Supplier IDs extracted from the file.
        '''
        e.g.                                            ID: 01, Name: Supplier1, Contact: 1234567890
        #line.split(',')                              ['ID: 01', ' Name: Supplier1', ' Contact: 1234567890']
        #line.split(',')[0]                            'ID: 01'
        #line.split(',')[0].split(":")             ['ID', ' 01']
        #line.split(',')[0].split(":")[1]                 ' 01'
        #line.split(',')[0].split(":")[1].strip()          '01'
        '''
        
    ### Start Collecting Supplier Information    
    print("\nNow you may start to Add a new Supplier")
    """
    Add a new Supplier ID
    """
    while True: # Loop until valid, if information Invalid: WON'T break the loop
        sup_id = input("Enter a 2-digit Supplier ID (e.g. 01, 89): ")
        # Check if the Supplier ID is valid (numeric and 2-digit)
        if not sup_id.isdigit() or len(sup_id) != 2:
            print("Error: Invalid Supplier ID. Please enter a VALID 2-digit Supplier ID.")
            continue  # Continue asking for a Valid Supplier ID
        else:
            # Check if the Supplier ID already exists in the file
            if sup_id in existing_id:
                print(f"Error: Supplier ID {sup_id} already exists. Please enter a unique Supplier ID.")
                continue  # Continue asking for a Valid Supplier ID
            break  # Supplier ID: Valid and unique, break this loop and continuing Adding a new Supplier Name
        
    """
    Add a new Supplier Name
    """
    while True: # Loop until valid, if information Invalid: WON'T break the loop
        sup_name = input("Enter the Supplier Name: ")
        # Check if Supplier Name is empty
        if not sup_name.strip():  
            print("Error: Invalid Supplier Name. Supplier Name cannot be empty.")
            continue  # Continue asking for a Valid Name
        else:
            break # Supplier Name: Valid, break this Loop and continuing writing informations to the suppliers.txt file

    """
    Add a new Supplier Contact Details
    """
    while True: # Loop until valid, if information Invalid: WON'T break the loop
        sup_contact = input("Enter the Supplier Contact Details(Only numbers): ")
        # Remove whitespaces
        sup_contact = sup_contact.replace(" ", "")
        # Check if the supplier Contact Details is valid (numeric, not empty, 7~11 digits)
        if not sup_contact.isdigit():
            print("Error: Invalid Contact Details. Please enter a Valid Contact Number( Only numbers ).")
            continue  # Continue asking for a valid contact
        elif not 6 < len(sup_contact) < 12:
            print("Error: Invalid Contact Details. Please enter a Valid Contact Number( 7~11 digits).")
            continue  # Continue asking for a valid contact  
        else:
            break # Contact Details: Valid, break this Loop and continuing writing informations to the suppliers.txt file

    """
    Once all the informations are valid, write it to the suppliers.txt file
    """
    with open('suppliers.txt', 'a') as sup_file:
        sup_file.write(f"ID: {sup_id} , Name: {sup_name} , Contact: {sup_contact}\n")
        
    """
    Confirmed that the Supplier Information was successfully added 
    """
    print(f"\nSupplier {sup_name} ID({sup_id}) was added successfully.")

    """
    Allow users to display supplier informations existed in suppliers.txt file(yes/no):
    """
    
"""
7. Place Order
"""
import datetime # Import datetime module
def place_order():
    while True:
        order_id = input("Enter Order ID:") # Get order ID as a string
        if not order_id.isdigit():
            print("Wrong Order ID Format, Please Try Again")
        elif not order_id.strip(): # Check if input is empty
            print("Please Enter An Order ID")
        else:
            break
    products =read_data("products.txt")#read the data from product file
    while True:
        product_id = input("Enter Product ID:") # Get product ID as a string
        result = [product for product in products if product[0] == product_id]
        if not result:
            print("Product Not Found, Please Try Again")
        elif not product_id.isdigit(): # Corrected to check product_id instead of order_id
            print("Wrong Product ID Format, Please Try Again")
        elif not product_id.strip(): # Check if input is empty
            print("Please Enter A Product ID")
        else:
            break # Exit the loop when a valid ID is found
    
    while True:
        quantity = input("Enter Product Quantity:") # Get quantity as a string
        if not quantity.isdigit(): # Check if quantity is a digit
            print("Product quantity should be in digits, Please Try Again.")
        elif not quantity.strip(): # Check if input is empty
            print("Please Enter Product Quantity")
        else:
            break

    while True:
        date = input("Enter Your Order Date (DD/MM/YYYY):") # Get date as a string
        try:
            datetime.datetime.strptime(date, "%d/%m/%Y") # Validate date format
            if not date.strip(): # Check if input is empty
                print("Please Enter Order Date")
            else:
                break
        except ValueError: # Catch the exception if the date format is incorrect
            print("Invalid Order Date, Please Try Again.")
    
    try:
        with open("orders.txt", "a") as f: # Open the file in append mode
            f.write(order_id + "," + product_id + "," + quantity + "," + date + "\n")
        return "Order placed successfully" # Return success message
    except Exception as e: # Handle any exceptions that may occur during file writing
        return f"An error occurred: {e}"

"""
8. View Inventory
"""
def view_inventory(product):
    try:
        with open(product, "r") as file:
            print("\n--- Inventory ---")
            print(f"{'ID':<10}{'Name':<20}{'Description':<30}{'Price':<10}")
            print("-" * 70)
        
        for line in file:
            # Split the line into product details
            product = line.strip().split(",")
            product_id, name, description, price = product
            # Print product details in a formaved way
            print(f"{product_id:<10}{name:<20}{description:<30}{price:<10}")
            
    except FileNotFoundError:
        print(f"Error: The file '{product}' was not found.")

"""
9. Parse Price
"""
PRODUCTS_FILE = 'products.txt'
SUPPLIERS_FILE = 'suppliers.txt'
ORDERS_FILE = 'orders.txt'

def parse_price(price_str):
    """Remove RM prefix and convert price to float."""
    try:
        return float(price_str.replace('RM', '').strip())
    except ValueError:
        print(f"Invalid price format: {price_str}")
        return 0.0


"""
10. Generate Reports
"""
def generate_reports():
    while True:
        print('\nReport Generation Menu')
        print('[1] Generate Low Stock Report')
        print('[2] Generate Product Sales Report')
        print('[3] Generate Supplier Orders Report')
        print('[4] Return to Main Menu')
        
        try:
            selection = int(input('Enter your selection: '))
            if selection == 1:
                low_stock_threshold = int(input('Enter the low stock threshold: '))
                print('\nLow Stock Report')
                print('Product ID, Name, Description, Price, Stock Quantity')
                print('------------------------------------------------------------')
                try:
                    with open(PRODUCTS_FILE, 'r') as f:
                        product_data = f.readlines()
                    for line in product_data:
                        product_details = line.strip().split(',')
                        if len(product_details) < 5:
                            print(f"Skipping invalid line: {line.strip()}")
                            continue
                        product_id, name, desc, price, stock_quantity = product_details
                        if int(stock_quantity) < low_stock_threshold:
                            print(f'{product_id}, {name}, {desc}, {price}, {stock_quantity}')
                except IOError as e:
                    print(f"Error reading {PRODUCTS_FILE}: {e}")
            
            elif selection == 2:
                print('\nProduct Sales Report')
                print('Order ID, Product ID, Product Name, Quantity, Total Earnings')
                print('------------------------------------------------------------')
                try:
                    with open(ORDERS_FILE, 'r') as f:
                        sales_data = f.readlines()
                    with open(PRODUCTS_FILE, 'r') as f:
                        product_lines = [line.strip().split(',') for line in f]
                    product_sales = {}
                    for line in sales_data:
                        sale_details = line.strip().split(',')
                        if len(sale_details) < 4:
                            print(f"Skipping invalid line: {line.strip()}")
                            continue
                        order_id, product_id, quantity, _ = sale_details
                        quantity = int(quantity)
                        product_name = next((p[1] for p in product_lines if p[0] == product_id), 'Unknown')
                        product_price = parse_price(next((p[3] for p in product_lines if p[0] == product_id), 'RM0'))
                        
                        if product_id in product_sales:
                            product_sales[product_id][1] += quantity
                        else:
                            product_sales[product_id] = [product_name, quantity]
                            
                    for product_id, details in product_sales.items():
                        total_earnings = details[1] * parse_price(next((p[3] for p in product_lines if p[0] == product_id), 'RM0'))
                        print(f'{product_id}, {details[0]}, {details[1]}, {total_earnings:.2f}')
                except IOError as e:
                        print(f"Error reading {ORDERS_FILE}: {e}")
                
            elif selection == 3:
                print('\nSupplier Orders Report')
                print('Supplier ID, Supplier Name, Contact, Order Count, Orders')
                print('------------------------------------------------------------')
        
                supplier_orders = {}

                try:
                    with open(ORDERS_FILE, 'r') as f:
                        order_data = [line.strip().split(',') for line in f.readlines()]
                        
                    with open(SUPPLIERS_FILE, 'r') as f:
                        suppliers = {}
                        for line in f:
                            parts = line.strip().split(',')
                            supplier_id = parts[0].split(':')[1].strip()
                            supplier_name = parts[1].split(':')[1].strip()
                            supplier_contact = parts[2].split(':')[1].strip()
                            suppliers[supplier_id] = [supplier_name, supplier_contact]
                    
                    for order in order_data:
                        if len(order) < 4:
                            print(f"Skipping invalid line: {','.join(order)}")
                            continue
                        
                        supplier_id = order[0]
                        product_id = order[1]
                        quantity = int(order[2])
                        order_date = order[3]
                        
                        if supplier_id not in suppliers:
                            print(f"Supplier ID {supplier_id} not found in suppliers file. Skipping...")
                            continue
                        supplier_name, supplier_contact = suppliers[supplier_id]

                        if supplier_id in supplier_orders:
                            supplier_orders[supplier_id][2] += 1
                            supplier_orders[supplier_id][3].append(
                            f'{product_id} (Qty: {quantity}, Date: {order_date})'
                            )
                        else:
                            supplier_orders[supplier_id] = [
                            supplier_name,
                            supplier_contact,
                            1,
                            [f'{product_id} (Qty: {quantity}, Date: {order_date})']
                            ]
                            
                        for supplier_id, details in supplier_orders.items():
                            print(f'{supplier_id}, {details[0]}, {details[1]}, {details[2]} Orders')
                            for order in details[3]:
                                print(f' {order}')
                                
                except IOError as e:
                    print(f"Error reading files: {e}")
            
            elif selection == 4:
                break
            else:
                print("Invalid selection. Choose 1 to 4.")
                
        except ValueError as e:
            print("Invalid input. Please enter a number between 1 and 4.")

"""
11. Main Program
"""
def main():
    while True:
        menu()
        choice = input("Select an option from 1 to 7: ")
        
        if choice == '1':
            add_products()
        elif choice == '2':
            update_products()
        elif choice == '3':
            add_supplier()
        elif choice == '4':
            place_order()
        elif choice == '5':
            view_inventory("products.txt")
        elif choice == '6':
            generate_reports()
        elif choice == '7':
            print("Program shuxng down, goodbye!")
            break
        else:
            print("Invalid choice: Choose from 1 to 7 only.")
    
main()
```
